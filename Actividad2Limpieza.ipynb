```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('inmuebles2.csv') 
data
```

```python
#Eliminar primera fila o filas innecesarias
data1= pd.read_csv('inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data3=data2.copy()
```

```python
#Sustituir valores nulos con promedio o media
data3["Superficie"]=data3["Superficie"].fillna(round(data2["Superficie"].mean(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].median(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Rellena con los valores de la fila sucesora
#Filtro por columnas
#Actividad_total=data3.iloc[ : , [0]]
#Actividad_total.fillna(method="bfill")
#data3["Fecha Alta"].fillna(method="bfill")
data3['Fecha Alta'] = data3['Fecha Alta'].fillna('2004-01-01 00:00:00')
#Revisamos que nuestro metodo haya funcionado
valores_nulos=data3.isnull().sum()
valores_nulos
#Concatenación de columnas
fecha = data3["Fecha Alta"].fillna('2004-01-01 00:00:00')
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Tipo'] = data3 ['Tipo'].fillna('Piso')
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Operación'] = data3 ['Operación'].fillna(method='ffill')
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Provincia'] = data3 ['Provincia'].fillna(method='bfill')
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Convertir archivo filtrado a CSV
data3.to_csv("DF sin NA.csv")
```
